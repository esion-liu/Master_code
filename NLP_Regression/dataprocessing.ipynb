This file is going to preprocess the game review dataset. Including:
keep only english reviews
make some static summary of reviews.

```python
import pandas as pd
from langdetect import detect, LangDetectException
import glob
import multiprocessing
```

```python
# Check if a review is written in English or not. Also check if the review is too short (less than 5 embeddings)

def is_english_and_long(text):
    try:
        # Convert to string to ensure compatibility with strip and detect
        text_str = str(text).strip()
        return detect(text_str) == 'en' and len(text_str.split()) >= 5
    except LangDetectException:
        return False
```

```python
csv_files = glob.glob('Dataset/*.csv')
csv_files.remove('Dataset\game_info.csv')
for file in csv_files:
    print(file)
```

```python
raise Exception('Already executed')
for file in csv_files:
    df = pd.read_csv(file)
    print('Going to process file ', file, ' Origin length', len(df))
    filtered_df = df[df['review'].apply(is_english_and_long)]
    filtered_df['review'] = filtered_df['review'].apply(lambda x: str(x).strip())
    filtered_df.to_csv(file, index=False)
    print('Process finished, file saved to ', file, ' length after processing is: ', len(filtered_df), '\n')
```

```python
# drive some static info
csv_files.remove('Dataset\game_review_all.csv')
dfs = []
lengths = []
for file in csv_files:
    df = pd.read_csv(file)
    length = len(df)
    dfs.append(df)
    lengths.append(length)
lengths
```

```python
from scipy.stats import norm
```

```python
df = pd.read_csv('Dataset\\Action_critic.csv')
data = df['score'].to_list()

# Create histogram data with bins for each number,
bins = np.arange(0, 105, 5)  # Bins from 0 to 100, incrementing by 5
hist, bin_edges = np.histogram(data, bins=bins)

# Plotting the histogram as a bar graph
plt.figure(figsize=(10, 6))
plt.bar(bin_edges[:-1], hist, width=4.8, align='edge', alpha=0.7, color='blue', label='Occurrence')

# Fit a Gaussian distribution to the data
mu, std = norm.fit(data)  # Get the parameters of the Gaussian fit
xmin, xmax = plt.xlim(0, 100)  # Limit x-axis for better fit visualization
x = np.linspace(xmin, xmax, 300)
p = norm.pdf(x, mu, std)

# Plot the Gaussian distribution
plt.plot(x, p * max(hist)/max(p), 'k', linewidth=2, label='Gaussian fit')

# Set labels and title
plt.xlabel('Rating Range')
plt.ylabel('Frequency')
plt.title('Action Game Rating Frequency Distribution')
plt.xticks(ticks=bins)  # Set x-ticks to match the bins
plt.legend()


plt.show()
```
