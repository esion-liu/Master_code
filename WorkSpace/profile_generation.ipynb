```python
# this notebook is used to generate the profile of first 10000 users in the user list
import psycopg2
import pickle
from Aggregation import personal_perf
from Aggregation import sampling
```

```python
# load user list
with open('data/user_list.pickle', 'rb') as f:
    user_list = pickle.load(f)
```

```python
postgres_conn_param = {
    'host': 'localhost',
    'port': '5432',
    'user': 'postgres',
    'password': 'Lyc199412',
    'database': 'steam'
}

conn = psycopg2.connect(**postgres_conn_param)
```

```python
profile = []
for j in range(10000):
    SQL = f"""
    SELECT gamea_name, gameb_name, prefered_game
    FROM llama3_pair
    WHERE reviewer_name = '{user_list[j]}'
    AND prefered_game != 'CannotJudge'
    """
    cursor = conn.cursor()
    cursor.execute(SQL)
    personal_pairs = cursor.fetchall()
    cursor.close()

    pref_pairs = []
    for i in personal_pairs:
        if i[0] == i[2]:
            pref_pairs.append((i[0], i[1]))
        elif i[1] == i[2]:
            pref_pairs.append((i[1], i[0]))
    
    pref = personal_perf.prefernce_generation(pref_pairs)
    pref_index = sampling.index_convertion(pref)

    profile.append(pref_index)
    if j%50 == 0:
        print(j)
```

```python
# Since StatRank library in R environment's index start from 1 instead of 0 all index need plus 1
for i in range(10000):
    for j in range(len(profile[i])):
        profile[i][j] += 1
profile[9998:]
```

```python
#save to csv
import csv
max_columns = 2838 # total game numner (use as alternatives)

with open("data/StatRankData.csv", mode='w', newline="") as f:
    writer = csv.writer(f)
    # write header
    writer.writerow([f"V{i+1}" for i in range(max_columns)])

    # write rows
    for row in profile:
        writer.writerow(row + [0]*(max_columns - len(row)))
```
